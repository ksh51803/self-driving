```python
# bias 없음
#x_trains = [1,   1.2, 2.3, 3, 3.2, 4.2, 4.6,   4, 5, 4.2]
xs_trains = [[1.0, 1],[1.2, 2],[2.3, 1],[3.0, 4],[3.2, 0],[4.2, 0],[4.6, 0],[4.0, 3],[5.0, 1],[4.2, 3]]
y_trains =  [ 2.4,     3.0,     4.4,     7.0,     5.2,     6.4,     7.1,     8.2,     9.0,     7.7]

m = len(y_trains)
ws = [0, 0]
x_tests = [[1.7, 2], [3.7, 1], [5.5, 3]]
```

```python
import numpy as np

xs_trains = np.array(xs_trains)
y_trains = np.array(y_trains)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs_trains[:, 0], xs_trains[:, 1], y_trains, 'o')
```

```python
# hypothesis(x1, x2) = w1*x1 + w2*x2
# hypothesis(xs) = w1*x1 + w2*x2
def h(xs):
    global ws
    #hypothesis = ws * xs
    hypothesis = np.dot(ws, xs)
    return hypothesis
```

```python
def cost(weights):      
    global ws
    ws = weights
    cost = np.sum([(h(xs) - y)**2 for xs, y in zip(xs_trains, y_trains)]) / m
    return cost
```

```python
from itertools import combinations
ws = list(range(-5, 10))
wss = np.array(list(combinations(ws, 2)))
```
