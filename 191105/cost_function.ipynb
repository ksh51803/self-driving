```python
# bias 없음
x_trains = [1,   1.2, 2.3, 3, 3.2, 4.2, 4.6,   4, 5, 4.2]
y_trains = [2.4, 3.0, 4.4, 7, 5.2, 6.4, 7.1, 8.2, 9, 7.7]
m = len(y_trains)
w = 0
x_tests = [1.7, 5.5, 12]
```

```python
# train set 분포 그래프
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x_trains, y_trains, 's')
```

```python
def h(x):
    global w
    hypothesis = w*x
    return hypothesis
```

```python
def cost(weight):      
    global w
    w = weight
    cost = sum([(h(x) - y)**2 for x, y in zip(x_trains, y_trains)]) / m
    return cost
```

```python
ws = range(-2, 6)
costs = [cost(w) for w in ws]
plt.plot(ws, costs, 'o--')
```
